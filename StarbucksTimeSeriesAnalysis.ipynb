# Starbucks Time Series Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from datetime import datetime
```

```python
df = pd.read_csv('starbucks.csv', index_col= 'Date', parse_dates=True)
```

```python
df['Close'].plot(figsize=(12,5));
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window= 60).mean()['Close'].plot()
```

The above plot shows us the trend of the closing price of the stocks of Starbucks. We find it is increaseing till 2016-01 and then decreasing till 2017-01 and then again increasing till 2017 July. It then faces a decline in the closing price and then remaining constant for the next 1 year and then declining till 2018 Dec and thereafter its increasing at a faster pace.
